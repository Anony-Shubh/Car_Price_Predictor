```python
from flask import Flask, render_template, request, redirect
from flask_cors import CORS, cross_origin
import pickle
import pandas as pd
import numpy as np

app = Flask(__name__)
cors = CORS(app)

# Load the pre-trained model
model = pickle.load(open('LinearRegressionModel.pkl', 'rb'))

# Load and preprocess the dataset
car = pd.read_csv('Cleaned_Car_data.csv')
# Handling imbalanced data should be done here in your data preparation pipeline before model training

@app.route('/', methods=['GET', 'POST'])
def index():
    companies = sorted(car['company'].unique())
    car_models = sorted(car['name'].unique())
    years = sorted(car['year'].unique(), reverse=True)
    fuel_types = car['fuel_type'].unique()
    transmissions = car['transmission'].unique()  # Assuming 'transmission' column exists

    # Insert a default option for each dropdown menu
    companies.insert(0, 'Select Company')
    car_models.insert(0, 'Select Model')
    years.insert(0, 'Select Year')
    fuel_types = np.insert(fuel_types, 0, 'Select Fuel Type')
    transmissions = np.insert(transmissions, 0, 'Select Transmission')

    return render_template('index.html', companies=companies, car_models=car_models, years=years, fuel_types=fuel_types, transmissions=transmissions)

@app.route('/predict', methods=['POST'])
@cross_origin()
def predict():
    company = request.form.get('company')
    car_model = request.form.get('car_models')
    year = int(request.form.get('year'))
    fuel_type = request.form.get('fuel_type')
    driven = int(request.form.get('kilo_driven'))
    transmission = request.form.get('transmission')

    # Define the features as per your trained model
    FEATURES = ['name', 'company', 'year', 'kms_driven', 'fuel_type', 'transmission']

    # Create a DataFrame for input data
    input_data = pd.DataFrame([[car_model, company, year, driven, fuel_type, transmission]], 
                              columns=FEATURES)

    # Predicting the price
    prediction = model.predict(input_data)
    print(prediction)

    return str(np.round(prediction[0], 2))

if __name__ == '__main__':
    app.run(debug=False, use_reloader=False)
```
